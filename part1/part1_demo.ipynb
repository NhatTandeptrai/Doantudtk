## Test Case 1: Cài đặt Gauss cơ bản & Partial Pivoting

```python
A1 = [
    [0, 2, 1],
    [1, -2, -3],
    [-1, 1, 2]
]
b1 = [-8, 0, 3]

print("Ma trận gốc A1:")
for row in A1: print(row)
print("Vector b1:", b1)

U1, c1, x1, swaps1 = gaussian_eliminate(A1, b1)
print(f"\n=> Nghiệm x1: {x1}")
print(f"=> Số lần hoán đổi dòng: {swaps1} lần")
print("=> Kiểm chứng với NumPy:")
verify_solution(A1, x1, b1)
```

## Test Case 2: Tính định thức

```python
# 2.1 Ma trận bình thường
A2 = [
    [2, -1, 1],
    [3, 3, 9],
    [3, 3, 5]
]
det_val = determinant(A2)
det_np = np.linalg.det(A2)
print("Ma trận A2_normal:")
for row in A2: print(row)
print(f"=> Định thức tự tính : {det_val}")
print(f"=> Định thức NumPy   : {det_np:.4f}")

# 2.2 Ma trận suy biến
A2_singular = [
    [1, 2, 3],
    [2, 4, 6],
    [7, 8, 9]
]
det_sing = determinant(A2_singular)
print(f"\nMa trận suy biến (hàng 2 = 2 * hàng 1):")
print(f"=> Định thức tự tính : {det_sing}")
print(f"=> Định thức NumPy   : {np.linalg.det(A2_singular):.4f}")
```

## Test Case 3: Ma trận nghịch đảo & Kiểm thử A * A^-1 = I

```python
A3 = [
    [4, 7],
    [2, 6]
]
inv_A3 = inverse(A3)

print("Ma trận A3:")
for row in A3: print(row)

print("\nMa trận nghịch đảo A3^-1:")
for row in inv_A3:
    print([round(val, 4) for val in row])

print("\nKiểm tra A3 * A3^-1 có ra ma trận đơn vị (I) không?")
identity_check = multiply(A3, inv_A3)
for row in identity_check:
    print([round(val, 4) for val in row])
```

## Test Case 4: Hạng và cơ sở các không gian

```python
A4 = [
    [1, 2, 0, -1],
    [2, 4, 1,  1],
    [3, 6, 1,  0]
]

print("Ma trận A4:")
for row in A4: print(row)

rank, row_space, col_space, null_space = rank_and_basis(A4)
print(f"\n=> Hạng của ma trận: {rank}")
print(f"=> Cơ sở không gian dòng:")
for r in row_space: print("  ", [round(v, 4) for v in r])

print(f"=> Cơ sở không gian cột:")
for c in col_space: print("  ", [round(v, 4) for v in c])

print(f"=> Cơ sở không gian nghiệm:")
for n_vec in null_space: print("  ", [round(v, 4) for v in n_vec])
```

## Test Case 5: Kiểm chứng kết quả bằng NumPy

```python
# Kiểm chứng tổng hợp với ma trận lớn hơn
A5 = [
    [2, 1, -1, 3],
    [4, 4, -3, 8],
    [-2, 5, 2, 1],
    [6, 3, -4, 12]
]
b5 = [5, 13, 0, 16]

print("Ma trận A5 (4x4):")
for row in A5: print(row)
print("Vector b5:", b5)

U5, c5, x5, swaps5 = gaussian_eliminate(A5, b5)
print(f"\n=> Nghiệm x5: {[round(v, 6) for v in x5]}")
print(f"=> Số lần hoán đổi: {swaps5}")

# So sánh với NumPy
x_np = np.linalg.solve(A5, b5)
print(f"=> Nghiệm NumPy  : {x_np}")

print("\n=> Kiểm chứng Ax = b:")
verify_solution(A5, x5, b5)

# Kiểm chứng định thức
det5 = determinant(A5)
det5_np = np.linalg.det(A5)
print(f"\n=> Định thức tự tính: {det5}")
print(f"=> Định thức NumPy  : {det5_np:.4f}")

# Kiểm chứng ma trận nghịch đảo
inv5 = inverse(A5)
inv5_np = np.linalg.inv(A5)
print(f"\n=> Sai số nghịch đảo (norm): {np.linalg.norm(np.array(inv5) - inv5_np):.2e}")
```
